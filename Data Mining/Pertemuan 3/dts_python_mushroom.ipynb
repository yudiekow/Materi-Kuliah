```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
#Import data latih
mush_train = pd.read_csv('mushrooms_train.csv')
```

```python
mush_train.head()
```

```python
#Import data uji
mush_test = pd.read_csv('mushrooms_test.csv')
```

```python
mush_test.head(10)
```

```python
#Ubah data latih dari string ke numerik
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
mush_train_num = mush_train.apply(lb.fit_transform)
```

```python
mush_train_num
```

```python
#Pisah kan fitur data latih dengan target data latih
x_train = mush_train_num.drop(["class"], axis = 1)
y_train = mush_train_num["class"].values
```

```python
#Pisahkan fitur data uji dengan target data uji
x_test = mush_test.drop(["class"], axis = 1)
y_test = mush_test["class"].values
```

```python
#Ubah bentuk fitur data uji ke numerik
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
x_test = x_test.apply(lb.fit_transform)
```

```python
#Normalisasi fitur data latih dan data uji
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
x_train = st.fit_transform(x_train)
x_test = st.fit_transform(x_test)
```

```python
#Buat model klasifikasi
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
#Uji data
knn.fit(x_train, y_train)
```

```python
#Prediksikan seluruh data uji
predictions = knn.predict(x_test)
```

```python
#Buat dataframe baru bernama mush_test_hasil
#Dataframe ini digunakan untuk menyimpan hasil akhir
mush_test_hasil = mush_test
mush_test_hasil["class"] = predictions
mush_test_hasil
```

```python
#Pada dataframe mush_test_hasil, kelas masih berbentuk numerik 0 dan 1
#Remap kelas menggunakan dictionary
#Dengan ketentuan p : 1 dan e : 0
#Tetapi, sebelum itu ubah kolom class ke tipe "string"
mush_test_hasil['class'] = mush_test_hasil['class'].astype(str)
#Setelah kolom diubah ke tipe string
#Buat dictionarynya
dict = {'1' : 'p', '0' : 'e'}
#Setelah jadi, ganti isi kolom class sesuai dengan dictionary
mush_test_hasil['class']= mush_test_hasil['class'].map(dict) 
mush_test_hasil
```

```python
#Export hasil data uji
mush_test_hasil.to_csv(r'C:\Users\User\mush_test_hasil.csv', index = None, header=True)
```
